## Exercice 3 :
En utilisant les fonctions données ci-dessus, proposer une version itérative et une version récursive pour la
fonction longueur(liste) qui renvoie la longueur de la liste chaînée liste telle que :

```python
>>> longueur(nil)
0
>>> longueur(cons(7, cons(0, cons(12, nil))))
3
```

```python
def longueur(liste):  # version itérative
    compteur = 0
    while not est_vide(liste):
        compteur += 1
        liste = queue(liste)
    return compteur
```
